#### Q5: TODO - What is the daily percentage change and analyse this company is stable to invest or not.
 #### r(t)=(Pt/P(t-1))-1

```python
Amazon_stock['returns']=(Amazon_stock['Close']/Amazon_stock['Close'].shift(1))-1
```

```python
Amazon_stock
```

```python
Amazon_stock.iloc[[Amazon_stock['returns'].argmax()]]
```

```python
Amazon_stock['returns'].hist(bins=100)
```

##### If the higher and wider  the histogram of the returns can be said as a high volatility, if compared to anyother stock. 

```python
Amazon_stock['returns'].plot(kind='kde')
```

##### cumilative return of the amazon stocks.

 It=(1+Rt)I(t-1)
 
 How much can we get if invest in past and by now hown much we can gain is simply we are doing.

```python
Amazon_stock['Cummulative_Return']=(1+Amazon_stock['returns']).cumprod()
Amazon_stock
```

#### If any investment made in amazon stocks before one year now the gain can be at a reate of +130%
